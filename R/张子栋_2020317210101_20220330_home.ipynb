(2) 利用各种方法进行检验是否服从正态分布？包括利用 QQ 图等进行可视化检验。

```r
data <- c(25, 45, 50, 54, 55, 61, 64, 68, 72, 75, 75,
          78, 79, 81, 83, 84, 84, 84, 85, 86, 86, 86,
          87, 89, 89, 89, 90, 91, 91, 92, 100)
# 正态性 W 检验方法
shapiro.test(data)

# ks 检验
ks.test(data, 'pnorm', mean(data), sd(data))

# Pearson 拟合优度卡方检验
chisq.test(data)

# QQ 图
qqnorm(data); qqline(data)
```
